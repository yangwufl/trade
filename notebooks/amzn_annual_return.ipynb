```python
import yfinance as yf
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_context("talk")
```

```python
amzn_mo_pricing = yf.download(
    tickers="AMZN",
    start=None,
    end="2021-05-05",
    interval="1mo",
)
```

```python
amzn_mo_pricing
```

```python
amzn_mo_pricing["Date"] = amzn_mo_pricing.index
amzn_mo_pricing["Month"] = amzn_mo_pricing.Date.apply(lambda d: d.month)
amzn_mo_pricing = amzn_mo_pricing[amzn_mo_pricing.Month==12].dropna(how="any")
```

```python
amzn_mo_pricing.Close.plot(figsize=(20,5));
```

```python
amzn_mo_pricing["Close_Prev"] = amzn_mo_pricing.Close.shift(1)
amzn_mo_pricing["Return"] = amzn_mo_pricing.Close/amzn_mo_pricing.Close_Prev - 1.0
amzn_mo_pricing
```

```python
amzn_mo_pricing.Return.hist(bins=10, histtype="step", grid=False, figsize=(20,5));
```

```python
amzn_mo_pricing.Return.mean(), amzn_mo_pricing.Return.median()
```

```python
amzn_mo_pricing.tail(10)
```

```python
amzn_mo_pricing.tail(10).Return.plot(figsize=(20,5));
```

```python
amzn_mo_pricing.tail(10).Return.mean(), amzn_mo_pricing.tail(10).Return.median()
```

```python
amzn_mo_pricing.tail(20).Return.mean(), amzn_mo_pricing.tail(20).Return.median()
```

```python
amzn_mo_pricing.tail(30).Return.mean(), amzn_mo_pricing.tail(30).Return.median()
```
